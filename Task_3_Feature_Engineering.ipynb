```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/clean_data_after_eda.csv")

# Step 1: Remove irrelevant columns
columns_to_drop = ['id', 'date_modif_prod']  # Example: Removing ID and modification date as they are not predictive
df.drop(columns=columns_to_drop, inplace=True)

# Step 2: Extract useful features from date columns
df['activation_year'] = pd.to_datetime(df['date_activ']).dt.year
df['renewal_month'] = pd.to_datetime(df['date_renewal']).dt.month
df['contract_length'] = (pd.to_datetime(df['date_end']) - pd.to_datetime(df['date_activ'])).dt.days

# Step 3: Create new features
df['consumption_trend'] = df['cons_last_month'] / (df['cons_12m'] + 1)  # Avoid division by zero
df['total_gas_and_elec'] = df['cons_12m'] + df['cons_gas_12m']  # Combining electricity and gas consumption

# Step 4: Save the updated dataset
df.to_csv("clean_data_with_new_features.csv", index=False)

print("Feature engineering completed. New dataset saved as clean_data_with_new_features.csv.")
```
