```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
client_data_path = "/content/client_data (1).csv"
price_data_path = "/content/price_data (1).csv"

client_df = pd.read_csv(client_data_path)
price_df = pd.read_csv(price_data_path)

# Get data types of each column
client_data_types = client_df.dtypes
price_data_types = price_df.dtypes

# Get descriptive statistics
client_desc_stats = client_df.describe()
price_desc_stats = price_df.describe()

# Plot distributions of numerical columns for client data
num_cols_client = client_df.select_dtypes(include=['number']).columns
fig, axes = plt.subplots(nrows=len(num_cols_client)//3 + 1, ncols=3, figsize=(15, 15))
axes = axes.flatten()

for i, col in enumerate(num_cols_client):
    sns.histplot(client_df[col].dropna(), bins=30, kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')

plt.tight_layout()
plt.show()

# Plot distributions of numerical columns for price data
num_cols_price = price_df.select_dtypes(include=['number']).columns
fig, axes = plt.subplots(nrows=len(num_cols_price)//3 + 1, ncols=3, figsize=(15, 15))
axes = axes.flatten()

for i, col in enumerate(num_cols_price):
    sns.histplot(price_df[col].dropna(), bins=30, kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')

plt.tight_layout()
plt.show()

# Return summary of data types and statistics
client_data_types, client_desc_stats, price_data_types, price_desc_stats
```
